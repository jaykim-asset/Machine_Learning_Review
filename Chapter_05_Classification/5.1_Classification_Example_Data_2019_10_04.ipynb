#### - If we look into the "Petal Length", we could distinguish "setosa" and "the others".  

```python
sns.distplot(df[df.species != 'setosa']['petal length (cm)'], hist=True, rug=True, label='setosa')
sns.distplot(df[df.species == 'setosa']['petal length (cm)'], hist=True, rug=True, label='others')
plt.legend()
plt.show()
```

- 2) Wine Data
    - target data set (y values)
        - 0, 1, 2 (three kinds of wines)
    - feature data set (x values)
        - 1) alcohol
        - 2) Malic acid
        - 3) Ash
        - 4) Alcalinity of ash
        - 5) Magnesium
        - 6) Total phenols
        - 7) Flavanoids
        - 8) Nonflavanoid phenols
        - 9) Proanthocyanins
        - 10) Color intensity
        - 11) Hue
        - 12) OD280/OD315 of diluted wines
        - 13) Proline

```python
from sklearn.datasets import load_wine
wine = load_wine()
print(wine.DESCR)
```

```python
df = pd.DataFrame(wine.data, columns=wine.feature_names)
sy = pd.Series(wine.target, dtype='category')
sy = sy.cat.rename_categories(wine.target_names)
df['class'] = sy
df.tail()
```

```python
sns.pairplot(vars=["alcohol", "alcalinity_of_ash", "total_phenols", "flavanoids"], 
             hue="class", data=df)
plt.show()
```
